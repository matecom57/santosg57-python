```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('iris.csv')

print(df.head())
plt.plot(df['species'], df['sepal_length'], 'o')
m1 = np.median(df['sepal_length'][df['species'] == 'setosa'])
m2 = np.median(df['sepal_length'][df['species'] == 'versicolor'])
m3 = np.median(df['sepal_length'][df['species'] == 'virginica'])
print([m1, m2, m3])
plt.plot(['setosa', 'versicolor', 'virginica'] ,[m1, m2, m3],'or')

plt.show()
```
