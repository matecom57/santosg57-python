```python
# https://es.wikipedia.org/wiki/Desarrollo_de_fractales_mediante_el_m%C3%A9todo_de_Mandelbrot

# Z = Z^4 + C

# tomamo un cuadrado formado por [-3, 3] y vemos sila itercion es finita.
# el cuadrado es ajustado en una matriz de tamaño 256 x 256

import numpy as np
import matplotlib.pyplot as plt

zz = np.arange(-3,3,.01)

img = np.zeros((256, 256,3))

for cx in zz:
    for cy in zz:
        c = complex(cx, cy)
        z = 0
        i = 0
        while i < 20:
            z = z**4 + c
            if abs(z) > 100000:
                break
            else:
                i = i+1
        if i <= 9:
            ix = int((cx+3)*256/6)
            iy = int((cy+3)*256/6)
            img[ix, iy,0] = i*255/9

plt.imshow(img)
plt.show()
```
